```python
# Take the corpus and remove the stop words
from gensim.parsing.preprocessing import remove_stopwords, \
                                         preprocess_string

```

```python
def remove_stopwords_corpus(corpus):

    new_corpus = {item: remove_stopwords(value) for item,value in corpus.items()}
        
    return new_corpus
```

```python
new_corpus = remove_stopwords_corpus(corpus_sonnets)
```
